```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
df = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
df.head()
```

```python
for i in df.values:
    sns.boxplot(i)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
scaled_data
```

```python
scaled_data.shape
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
x_pca.shape
```

```python
x_pca
```

```python
plt.figure(figsize=(30,15))
plt.scatter(x_pca[:,0],  # COl- 1
            x_pca[:,1], # Col-2
            c=cancer['target'],
            cmap='rainbow')
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```

```python
pca.components_
```

```python
map= pd.DataFrame(pca.components_,columns=cancer['feature_names'])
plt.figure(figsize=(30,15))
sns.heatmap(map,cmap='twilight')
```

```python
from sklearn.datasets import fetch_olivetti_faces
```

```python
Data = fetch_olivetti_faces()
```

```python
Data.keys()
```

```python
Data.images[0].shape
```

```python
plt.imshow(Data.images[99])
```

```python
Data.data.shape  # H(array) * W(row)* D(no col)
```

```python
from sklearn.decomposition import PCA
#n_components=0.80 means it will return the Eigenvectors that have the 80% of the variation in the dataset
faces_pca = PCA(n_components=0.8)
faces_pca.fit(Data.data)
```

```python
faces_pca.components_.shape
```

```python
components = faces_pca.transform(Data.data)
projected = faces_pca.inverse_transform(components)
```

```python
fig, axes = plt.subplots(2,10,
                         figsize=(9,3),
                         subplot_kw={'xticks':[], 'yticks':[]},
                         gridspec_kw=dict(hspace=0.01, wspace=0.01))
for i,ax in enumerate(axes.flat):
    ax.imshow(faces_pca.components_[i].reshape(64,64))
```

```python
components = faces_pca.transform(Data.data)
projected = faces_pca.inverse_transform(components)
fig, axes = plt.subplots(10,10,
                         figsize=(30,15), 
                         subplot_kw={'xticks':[], 'yticks':[]},
            gridspec_kw=dict(hspace=0.01, wspace=0.01))
for i, ax in enumerate(axes.flat):
    ax.imshow(projected[i].reshape(64,64),cmap="gray")
```

### Advantages of dimensionality reduction
1. It reduces the time and storage space required.
2. Removal of multi-collinearity improves the interpretation of the parameters of the machine learning model.
3. It becomes easier to visualize the data when reduced to very low dimensions such as 2D or 3D.
4. It avoids the curse of dimensionality.
